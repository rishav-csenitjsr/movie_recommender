```python
import pandas as pd
```

```python
da=pd.read_csv('train.csv')
```

```python
titanic_data = da.drop(['PassengerId', 'Name', 'Ticket'], axis=1)
titanic_data.head()
```

```python
import numpy as np
```

```python
ba= titanic_data.drop([  'Embarked'],axis=1)
X = np.array([ba['Pclass'],ba['Sex']=='male',ba['Age'].fillna(25)])


X=X.T

Y = np.array(ba['Survived'])
from sklearn.metrics import mean_squared_error
from sklearn.datasets import make_friedman1
from sklearn.ensemble import GradientBoostingRegressor


est = GradientBoostingRegressor(n_estimators=100, learning_rate=0.1,
    max_depth=1, random_state=0, loss='ls').fit(X,Y)
```

```python
t=pd.read_csv('test.csv')
```

```python
X = np.array([t['Pclass'],t['Sex']=='male',t['Age'].fillna(25)])
print (X.shape)
X=X.T
```
